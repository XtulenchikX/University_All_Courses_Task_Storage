### <center> _Чтение и предобработка тестовых данных_

```python
# Чтение тестовых данных
test_data = pd.read_excel('test_data.xlsx')
test_data.head()
```

```python
# Тестовые данные
target_variable_name = 'price'
features = test_data.drop(target_variable_name, axis=1).columns
X_test, y_test = test_data[features], test_data[target_variable_name]
```

### <center> _Валидация моделей на тестовой выборке и визуализация результатов_

```python
# Получение прогноза целевой переменной с помощью модели линейной регрессии
y_test_lr_pred = linear_regression_model.predict(X_test)

# Получение прогноза целевой переменной с помощью модели случайного леса
y_test_rf_pred = random_forest_model.predict(X_test)
```

```python
# Визуализация прогноза модели линейной регрессии и настоящие значения из тестовой выборки
plt.figure(figsize=(7, 7))
plt.scatter(y_test, y_test_lr_pred) # рисуем точки, соответствущие парам настоящее значение - прогноз
plt.plot([0, max(y_test)], [0, max(y_test_lr_pred)]) # рисуем прямую, на которой предсказания и настоящие значения совпадают
plt.xlabel('Настоящая цена', fontsize=20)
plt.ylabel('Предсказанная цена', fontsize=20)
plt.title('Модель линейной регрессии')
```

```python
# Визуализация прогноза модели случайного леса и настоящие значения из тестовой выборки
plt.figure(figsize=(7, 7))
plt.scatter(y_test, y_test_rf_pred) # рисуем точки, соответствущие парам настоящее значение - прогноз
plt.plot([0, max(y_test)], [0, max(y_test_rf_pred)]) # рисуем прямую, на которой предсказания и настоящие значения совпадают
plt.xlabel('Настоящая цена', fontsize=20)
plt.ylabel('Предсказанная цена', fontsize=20)
plt.title('Модель случайного леса')
```

```python
# Функция вычисления метрик MAE, RMSE, R2
def print_metrics(y, y_pred):
    print('Train MAE: {:.7f}'.format(mean_absolute_error(y, y_pred)))
    print('Train RMSE: {:.7f}'.format(mean_squared_error(y, y_pred, squared=False)))
    print('Train R^2: {:.7f}'.format(r2_score(y, y_pred)))
```

```python
# Метрики для модели линейной регрессии
print('Linear regression model:')
print_metrics(y_test, y_test_lr_pred)

# Метрики для модели случайного леса
print('Random forest model:')
print_metrics(y_test, y_test_rf_pred)
```

Исходя из полученных результатов вычисления метрик и построенных графиков видно, что модель случайного леса работает лучше модели линейной регрессии. Одной из главных причин возникновения такого явления являются выбросы данных (неадекватные данные, то есть слишком низкая или слишком высокая цена, например в нашей тестовой выборке есть цена = 1), которые влияют на модель линейной регрессии намного сильнее, чем на модель случайного леса.

### <center> _Кодирование категориальных признаков в данные с помощью LabelEncoder_

```python
# Повторная загрузка данных
training_data = pd.read_excel('training_data.xlsx')
training_data = training_data.drop('Unnamed: 0', axis = 1)
training_data = training_data.dropna()

test_data = pd.read_excel('test_data.xlsx')
test_data = test_data.drop('Unnamed: 0', axis = 1)
test_data = test_data.dropna()
```

```python
training_data.head(10)
```

```python
test_data.head(10)
```

```python
# Кодирование категориальных признаков
text_categor_cols = ['title_status', 'transmission', 'drive', 'size']

le = LabelEncoder()

for col in text_categor_cols:
    training_data[col] = le.fit_transform(training_data[col]) + 1
    test_data[col] = le.transform(test_data[col]) + 1

```

```python
training_data.head(10)
```

```python
test_data.head(10)
```

```python
target_variable_name = 'price'

# Тренировочные данные с закодированными категориальными признаками с помощью LabelEncoder
features = training_data.drop(target_variable_name, axis=1).columns
X_train_le, y_train_le = training_data[features], training_data[target_variable_name]

# Тестовые данные с закодированными категориальными признаками с помощью LabelEncoder
features = test_data.drop(target_variable_name, axis=1).columns
X_test_le, y_test_le = test_data[features], test_data[target_variable_name]
```

```python
# Обучение модели
random_forest_model.fit(X_train_le, y_train_le)

# Получение прогноза целевой переменной
y_test_rf_pred_le = random_forest_model.predict(X_test_le)
```

```python
# Вычисление метрик с использование категориальных признаков
print('With categor cols, LabelEncoder:')
print_metrics(y_test_le, y_test_rf_pred_le)

# Вычисление метрик без категориальных признаков
print('Without categor cols:')
print_metrics(y_test, y_test_rf_pred)
```

```python
# Визуализация результатов с использованием категориальных признаков и без них
plt.figure(figsize=(7, 7))
plt.scatter(y_test_le, y_test_rf_pred_le, alpha=0.9, label='Числовые + текстовые признаки', s=10)
plt.scatter(y_test, y_test_rf_pred, color='orange', alpha=0.4, label='Только числовые признаки', s=10)
plt.plot([0, max(y_test_le)], [0, max(y_test_rf_pred_le)])

plt.legend()
plt.xlabel('Настоящая цена', fontsize=20)
plt.ylabel('Предсказанная цена', fontsize=20)
```

### <center> _Кодирование категориальных признаков в данные с помощью One Hot Encoding_

```python
# Повторная загрузка данных
training_data = pd.read_excel('training_data.xlsx')
training_data = training_data.drop('Unnamed: 0', axis = 1)
training_data = training_data.dropna()

test_data = pd.read_excel('test_data.xlsx')
test_data = test_data.drop('Unnamed: 0', axis = 1)
test_data = test_data.dropna()
```

```python
training_data.head(10)
```

```python
test_data.head(10)
```

```python
# Категориальные и остальные признаки
categor_cols = ['condition', 'cylinders', 'title_status', 'transmission', 'drive', 'size']
others_cols = ['price', 'year', 'odometer', 'lat', 'long', 'weather']
```

```python
# Создание прототипа OneHotEncoder
one_hot_encoder = OneHotEncoder(sparse_output=False)
```

```python
# Отделяем категориальные признаки от остальных у обучающей выборки и тестовой в отдельные таблички
categor_columns_training = training_data[categor_cols]
others_columns_training = training_data[others_cols]
categor_columns_test = test_data[categor_cols]
others_columns_test = test_data[others_cols]
```

```python
# Применяем fit_transform и transform
result_encoding_training = one_hot_encoder.fit_transform(categor_columns_training)
result_encoding_test = one_hot_encoder.transform(categor_columns_test)
```

```python
# Преобразуем результаты в табличку
result_encoding_training = pd.DataFrame(result_encoding_training, index=training_data.index)
result_encoding_test = pd.DataFrame(result_encoding_test, index=test_data.index)
```

```python
# Соединяем в одну табличку новые преобразованные признаки
training_data = pd.concat([others_columns_training, result_encoding_training], axis=1)
test_data = pd.concat([others_columns_test, result_encoding_test], axis=1)

# Преобразуем признаки к типу str
training_data.columns = training_data.columns.astype(str)
test_data.columns = test_data.columns.astype(str)
```

```python
training_data.head(10)
```

```python
test_data.head(10)
```

```python
target_variable_name = 'price'

# Тренировочные данные с закодированными категориальными признаками с помощью OneHotEncoder
features = training_data.drop(target_variable_name, axis=1).columns
X_train_ohe, y_train_ohe = training_data[features], training_data[target_variable_name]

# Тестовые данные с закодированными категориальными признаками с помощью OneHotEncoder
features = test_data.drop(target_variable_name, axis=1).columns
X_test_ohe, y_test_ohe = test_data[features], test_data[target_variable_name]
```

```python
# Обучение модели
random_forest_model.fit(X_train_ohe, y_train_ohe)

# Получение прогноза целевой переменной
y_test_rf_pred_ohe = random_forest_model.predict(X_test_ohe)
```

```python
# Вычисление метрик с использование категориальных признаков и OneHotEncoder
print('With categor cols, OneHotEncoder:')
print_metrics(y_test_ohe, y_test_rf_pred_ohe)
print('---------------')
# Вычисление метрик с использование категориальных признаков и LabelEncoder
print('With categor cols, LabelEncoder:')
print_metrics(y_test_le, y_test_rf_pred_le)
print('---------------')
# Вычисление метрик без категориальных признаков
print('Without categor cols:')
print_metrics(y_test, y_test_rf_pred)
```

```python
# Визуализация результатов с использованием категориальных признаков (LabelEncoder OneHotEncoder) и без них
plt.figure(figsize=(7, 7))
plt.scatter(y_test_le, y_test_rf_pred_le, color='blue', alpha=0.9, label='Числовые + LabelEncoder', s=5)
plt.scatter(y_test_ohe, y_test_rf_pred_ohe, color='green', alpha=0.9, label='Числовые + OneHotEncoder', s=5)
plt.scatter(y_test, y_test_rf_pred, color='orange', alpha=0.4, label='Только числовые признаки', s=5)
plt.plot([0, max(y_test_le)], [0, max(y_test_rf_pred_le)])

plt.legend()
plt.xlabel('Настоящая цена', fontsize=20)
plt.ylabel('Предсказанная цена', fontsize=20)
```
